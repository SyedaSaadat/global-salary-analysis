```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/SyedaSaadat/global-salary-analysis/main/adult.csv'
df = pd.read_csv(url)

df.head()
```

```python
# filter rows where country starts with 'S'
country_data = df[df['nativecountry'].str.startswith('S')]

# Preview the filtered data
country_data.head()
```

```python
numeric_attribute = 'incomeUSD'

min_value = country_data[numeric_attribute].min()
max_value = country_data[numeric_attribute].max()
range_value = max_value - min_value
mean_value = country_data[numeric_attribute].mean()
median_value = country_data[numeric_attribute].median()
mode_value = country_data[numeric_attribute].mode().values[0]
std_deviation = country_data[numeric_attribute].std()
q1_value = country_data[numeric_attribute].quantile(0.25)
q3_value = country_data[numeric_attribute].quantile(0.75)

print(f"Min: {min_value:.2f}")
print(f"Max: {max_value:.2f}")
print(f"Range: {range_value:.2f}")
print(f"Mean: {mean_value:.2f}")
print(f"Median: {median_value:.2f}")
print(f"Mode: {mode_value}")
print(f"Std Dev: {std_deviation:.2f}")
print(f"Q1: {q1_value:.2f}")
print(f"Q3: {q3_value:.2f}")

```
